```python
import pandas as pd 
import pickle

max_num_seeds = 90
#decided to use all emojis in the corpus, importance will already be affected by the frequency in the dataset
```

```python
tr_lexicon = pd.read_json("../data/lexicons/tr_emotion_lexicon.json")
tr_lexicon['emotion'] = tr_lexicon.idxmax(axis=1)
#filtering out only single word indexes
tr_lexicon = tr_lexicon[tr_lexicon.index.str.count(' ')==0]
#Filtering by discrete emotion
tr_anger = tr_lexicon[tr_lexicon.emotion=='Anger'][['Anger']]
tr_happy = tr_lexicon[tr_lexicon.emotion=='Happy'][['Happy']]
tr_sad = tr_lexicon[tr_lexicon.emotion=='Sadness'][['Sadness']]
tr_fear = tr_lexicon[tr_lexicon.emotion=='Fear'][['Fear']]
tr_disgust = tr_lexicon[tr_lexicon.emotion=='Disgust'][['Disgust']]
tr_surprise = tr_lexicon[tr_lexicon.emotion=='Surprise'][['Surprise']]

#Seed Selection top max_num_seeds
tr_anger_seeds = tr_anger.sort_values(by='Anger',ascending=False)[0:max_num_seeds].index.to_list()
tr_happy_seeds = tr_happy.sort_values(by='Happy',ascending=False)[0:max_num_seeds].index.to_list()
tr_sad_seeds = tr_sad.sort_values(by='Sadness',ascending=False)[0:max_num_seeds].index.to_list()
tr_fear_seeds = tr_fear.sort_values(by='Fear',ascending=False)[0:max_num_seeds].index.to_list()
tr_disgust_seeds = tr_disgust.sort_values(by='Disgust',ascending=False)[0:max_num_seeds].index.to_list()
tr_surprise_seeds = tr_surprise.sort_values(by='Surprise',ascending=False)[0:max_num_seeds].index.to_list()
```

```python
#lexicon is provided in files by emotion, we load them separately
en_anger = pd.read_csv('../data/lexicons/anger-scores.txt',sep='\t',index_col=0,header=None,names=['Score'])
en_sad = pd.read_csv('../data/lexicons/sadness-scores.txt',sep='\t',index_col=0,header=None,names=['Score'])
en_fear = pd.read_csv('../data/lexicons/fear-scores.txt',sep='\t',index_col=0,header=None,names=['Score'])
en_disgust = pd.read_csv('../data/lexicons/disgust-scores.txt',sep='\t',index_col=0,header=None,names=['Score'])
en_happy = pd.read_csv('../data/lexicons/joy-scores.txt',sep='\t',index_col=0,header=None,names=['Score'])
en_surprise = pd.read_csv('../data/lexicons/surprise-scores.txt',sep='\t',index_col=0,header=None,names=['Score'])

#merge the seperate lexicons in one dataframe

en_lexicon = en_anger.merge(en_sad,left_index=True,right_index=True,how='outer')\
    .merge(en_fear,left_index=True,right_index=True,how='outer')\
    .merge(en_disgust,left_index=True,right_index=True,how='outer')\
    .merge(en_happy,left_index=True,right_index=True,how='outer')\
    .merge(en_surprise,left_index=True,right_index=True,how='outer')

en_lexicon.columns = ['Anger','Sadness','Fear','Disgust','Happy','Surprise']

en_lexicon.fillna(0,inplace=True)

en_lexicon['emotion'] = en_lexicon.idxmax(axis=1)

#Filtering by discrete emotion
en_anger = en_lexicon[en_lexicon.emotion=='Anger'][['Anger']]
en_happy = en_lexicon[en_lexicon.emotion=='Happy'][['Happy']]
en_sad = en_lexicon[en_lexicon.emotion=='Sadness'][['Sadness']]
en_fear = en_lexicon[en_lexicon.emotion=='Fear'][['Fear']]
en_disgust = en_lexicon[en_lexicon.emotion=='Disgust'][['Disgust']]
en_surprise = en_lexicon[en_lexicon.emotion=='Surprise'][['Surprise']]

#Seed Selection top max_num_seeds

en_anger_seeds = sorted(en_anger.sort_values(by='Anger',ascending=False)[0:max_num_seeds].index.to_list())
en_happy_seeds = sorted(en_happy.sort_values(by='Happy',ascending=False)[0:max_num_seeds].index.to_list())
en_sad_seeds = sorted(en_sad.sort_values(by='Sadness',ascending=False)[0:max_num_seeds].index.to_list())
en_fear_seeds = sorted(en_fear.sort_values(by='Fear',ascending=False)[0:max_num_seeds].index.to_list())
en_disgust_seeds = sorted(en_disgust.sort_values(by='Disgust',ascending=False)[0:max_num_seeds].index.to_list())
en_surprise_seeds = sorted(en_surprise.sort_values(by='Surprise',ascending=False)[0:max_num_seeds].index.to_list())
```

```python
path_seeds = "/Users/atanaltay/Documents/Research/inwork/marketing_emotions/codes/data/all_marketing_data/seeds/"
with open(path_seeds+'tr90/anger', 'wb') as fp:
    pickle.dump(tr_anger_seeds, fp)
with open(path_seeds+'tr90/happy', 'wb') as fp:
    pickle.dump(tr_happy_seeds, fp)
with open(path_seeds+'tr90/sad', 'wb') as fp:
    pickle.dump(tr_sad_seeds, fp)
with open(path_seeds+'tr90/fear', 'wb') as fp:
    pickle.dump(tr_fear_seeds, fp)
with open(path_seeds+'tr90/disgust', 'wb') as fp:
    pickle.dump(tr_disgust_seeds, fp)
with open(path_seeds+'tr90/surprise', 'wb') as fp:
    pickle.dump(tr_surprise_seeds, fp)
```
